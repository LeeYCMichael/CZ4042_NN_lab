```python
# Import PyTorch
import torch
import torch.nn as nn
import numpy as np
```

## Define the RNNs

```python
class RNNPartA(nn.Module):
    def __init__(self):
        super(RNNPartA, self).__init__()

        # Specify the weights and biases
        self.u = nn.Parameter(torch.tensor([[-1.0, 0.5], [0.5, 0.1], [0.2, -2]]))
        self.w = nn.Parameter(torch.tensor([[2.0, 1.3], [1.5, 0.0]]))
        self.v = nn.Parameter(torch.tensor([[2.0], [-1.5]]))
        self.b = nn.Parameter(torch.tensor([0.2, 0.2]))
        self.c = nn.Parameter(torch.tensor([0.5]))

    def forward(self, x):
        h = torch.zeros(1, 2)  # initial hidden vector

        for i in range(len(x)):
            h = torch.tanh(torch.mm(x[i], self.u) + torch.mm(h, self.w) + self.b)
            y = torch.sigmoid(torch.mm(h, self.v) + self.c)

            # Print out the results during iteration
            print(f'h({i + 1}) = {h}')
            print(f'y({i + 1}) = {y}')
            print(' ')

        return y

class RNNPartB():
    def __init__(self):
        # Specify the weights and biases
        self.u = nn.Parameter(torch.tensor([[-1.0, 0.5], [0.5, 0.1], [0.2, -2]]))
        self.w = nn.Parameter(torch.tensor([[2.0, 1.3]]))
        self.v = nn.Parameter(torch.tensor([[2.0], [-1.5]]))
        self.b = nn.Parameter(torch.tensor([0.2, 0.2]))
        self.c = nn.Parameter(torch.tensor([0.5]))

    def __call__(self, x):
        y = torch.zeros(1, 1)
        for i in range(len(x)):
            h = torch.tanh(torch.mm(x[i], self.u) + torch.mm(y, self.w) + self.b)
            y = torch.sigmoid(torch.mm(h, self.v) + self.c)

            # Print out the results during iteration
            print(f'h({i + 1}) = {h}')
            print(f'y({i + 1}) = {y}')
            print(' ')

        return y
```

## Create the inputs


```python
# Initialize the inputs, and put them in a list
x1 = torch.tensor([[1, 2, -1]], dtype=torch.float32)
x2 = torch.tensor([[-1, 1, -2]], dtype=torch.float32)
x3 = torch.tensor([[0, 3, 1]], dtype=torch.float32)
x4 = torch.tensor([[2, -1, 0]], dtype=torch.float32)
x = [x1, x2, x3, x4]
```

## Initialize the RNNs and compute the outputs

You can verify the results here with your own calculations.

```python
# Intialize the RNNs we specified above
rnn_a = RNNPartA()
rnn_b = RNNPartB()

# Pass the inputs into the RNNs, and assign the final outputs to a variable y_a and y_b
print('---Part A---')
y_a = rnn_a(x)
print('---Part B---')
y_b = rnn_b(x)
```
