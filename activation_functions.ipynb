# Activation Functions

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
    
print(torch.__version__)
```

```python
# inputs for ploting activation functions
u = torch.arange(-10.0, 10.0, 0.1)
```

# Threshold activation

###  $thresh(u)=1(u>0)$

```python
def threshold(u):
	return (u>0.0).float()
```

```python
# thresh
y = threshold(u)
```

```python
plt.figure()
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.xticks([-5, 5])
plt.yticks([0.5, 1.0])
plt.plot(u.numpy(),y.numpy())
plt.xlabel('$u$')
plt.title('thresh($u$)')
plt.show
```

# Sigmoid activation

### $sigmoid(u) = \frac{1}{1+ e^{-u}}$

```python
# sigmoid
y = torch.sigmoid(u)
```

```python
plt.figure()
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.xticks([-5, 5])
plt.yticks([0.5, 1.0])
plt.plot(u.numpy(),y.numpy())
plt.xlabel('$u$')
plt.title('sigmoid($u$)')
plt.show
```

# Linear activation

### $f(u) = u$

```python
# linear
y = u
```

```python
plt.figure()
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.xticks([-5, 5])
plt.yticks([-5, 5])
plt.plot(u.numpy(),y.numpy())
# pylab.xlabel('$u$')
plt.title('linear($u$)')
plt.show
```

# ReLU activation

### $ReLU(u) = \max\{0, u\}$

```python
# relu
f = nn.ReLU()
y = f(u)
```

```python
plt.figure()
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.xticks([-5, 5])
plt.yticks([5])
plt.plot(u.numpy(),y.numpy())
plt.xlabel('$u$')
plt.title('ReLU($u$)')
plt.show
```

# Tanh activation

### $tanh(u) = \frac{e^u - e^{-u}}{e^u + e^{-u}}$

```python
# tanh
y = torch.tanh(u)
```

```python
plt.figure()
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.xticks([-5,  5])
plt.yticks([-1.0, -0.5, 0.5, 1.0])
plt.plot(u.numpy(),y.numpy())
# pylab.xlabel('$u$')
plt.title('tanh($u$)')
plt.show
```
